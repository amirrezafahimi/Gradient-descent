```python
# defining the function 
def f(x): 
    return x ** 3 + 5 * x ** 2 + 7 * x + 2

def df(x):
    return 3 * x ** 2 + 10 * x + 7
```

```python
def gradient_descent(derivative_func, initial_position, learning_rate, 
                     precision=.001, epochs=1000 ):
    
    new_value = initial_position
    previouse_value = new_value
    list_of_positions = [new_value]
    list_of_gradients = [derivative_func(new_value)]
    
    for number_of_iteration in range(epochs):    
        previouse_value = new_value
        gradient = derivative_func(previouse_value)
        new_value = previouse_value - learning_rate * gradient
        
        list_of_positions.append(new_value)
        list_of_gradients.append(derivative_func(new_value))
        
        stepsize = abs(new_value - previouse_value)
        
        if (stepsize < precision): break
    
    return np.array(list_of_positions), np.array(list_of_gradients), number_of_iteration, new_value
```

```python
list_x, gradient, iteration, the_minimum = gradient_descent(df, 1.8, .03)
```

```python
print('the number of iterations: ', iteration)
print('the minimum occurs at:', the_minimum)
```

```python
def k(x, y):
    r = 3**(-x**2 - y**2)
    return 1 / (r + 1)

def kpx(x, y):
    r = 3**(-x**2 - y**2)
    return 2*x*np.log(3)*r / (r + 1)**2

def kpy(x, y):
    r = 3**(-x**2 - y**2)
    return 2*y*np.log(3)*r / (r + 1)**2
```

```python
x_4 = np.linspace(-2, 2, 500)
y_4 = np.linspace(-2, 2, 500)

x_4, y_4 = np.meshgrid(x_4, y_4)
```

```python
fig = plt.figure(figsize=[16, 12])
ax = fig.gca(projection='3d')

ax.plot_surface(x_4, y_4, k(x_4, y_4), cmap=cm.coolwarm, alpha=.6)
```

```python
# Setup
multiplier = 0.1
max_iter = 200
params = np.array([1.8, 1.0]) # initial guess
values_array = params.reshape(1, 2)
print(values_array.shape)

for n in range(max_iter):
    gradient_x = kpx(params[0], params[1])
    gradient_y = kpy(params[0], params[1])
    gradients = np.array([gradient_x, gradient_y])
    params = params - multiplier * gradients
    #values_array = np.append(values_array, params.reshape(1, 2), axis=0)
    values_array = np.concatenate((values_array, params.reshape(1, 2)), axis=0)
    
    
# Results
print('Values in gradient array', gradients)
print('Minimum occurs at x value of: ', params[0])
print('Minimum occurs at y value of: ', params[1])
print('The cost is: ', k(params[0], params[1]))
```

```python
mean = [0, 0]
cov = [[1, 2], [2, 5]]
X = np.random.multivariate_normal(mean, cov, 100)
X.shape
```

```python
x_scatter = X[:, 0]
y_scatter = X[:, 1]
plt.scatter(X[:, 0], X[:, 1], s=10, color='blue')
```

```python
regr = LinearRegression()
regr.fit(X[:, np.newaxis, 0], X[:, np.newaxis, 1])
```

```python
print(regr.coef_[0, 0])
print(regr.intercept_[0])
```

```python
plt.scatter(X[:, 0], X[:, 1], s=10, color='blue')
plt.plot(x_scatter, regr.predict(X[:, np.newaxis, 0]), color='orange', lw=1, alpha=.5)
```

```python
predicted_y = regr.intercept_ + regr.coef_ * X[:, np.newaxis, 0]
```

```python
# Challenge: Write a python function mse(y, y_hat) returns the MSE? Call the mse(y, y_hat)
# function and print out the MSE for the y_hat calculated above. 

def mse(y, y_hat):
    mse_calc = np.average((y - y_hat)**2, axis=0)
    return mse_calc
```

```python
mse(X[:, np.newaxis, 1], predicted_y)[0]
```

```python
epoch = 300
theta_0 = np.linspace(-5, 5, epoch)
theta_1 = np.linspace(-5, 5, epoch)

theta_0, theta_1 = np.meshgrid(theta_0, theta_1)
```

```python
cost_value = np.empty((epoch, epoch))
```

```python
for i in range(epoch):
    for j in range(epoch):
        y_hat = theta_0[i, j] + theta_1[i, j] * X[:, np.newaxis, 1]
        cost_value[i, j] = mse(X[:, np.newaxis, 1], y_hat)
```

```python
fig = plt.figure(figsize=[16, 12])
ax = fig.gca(projection='3d')

ax.plot_surface(theta_0, theta_1, cost_value, cmap=cm.coolwarm, alpha=.6)
```

```python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size
    
    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)
    
    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
```

```python
multiplier = 0.1
thetas = np.array([5, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(X[:, np.newaxis, 0], thetas[0] + thetas[1]*X[:, np.newaxis, 1])

for i in range(1000):
    thetas = thetas - multiplier * grad(X[:, np.newaxis, 0], X[:, np.newaxis, 1], thetas)
    
    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mse(X[:, np.newaxis, 0], thetas[0] + thetas[1]*X[:, np.newaxis, 1]))
    
# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(X[:, np.newaxis, 0], thetas[0] + thetas[1]*X[:, np.newaxis, 1]))
```

```python
fig = plt.figure(figsize=[16, 12])
ax = fig.gca(projection='3d')

ax.plot_surface(theta_0, theta_1, cost_value, cmap=cm.coolwarm, alpha=.6)
ax.scatter(plot_vals[:, 0], plot_vals[:, 1], mse_vals, s=80, color='black')
```
